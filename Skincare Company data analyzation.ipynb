# Read Total profit of all months and show it using a line plot

```python
df_profit = df.groupby('month_number').sum()['total_profit'].reset_index()
```

```python
plt.figure(figsize=[15,5])
plt.plot(df_profit['month_number'],df_profit['total_profit'],color='#b80045')
plt.xticks(rotation = 'vertical',size=8)

plt.title('total profit month-wise')
plt.ylabel('Total_profit')
plt.xlabel('Month')
```

# What is the average profit of the company in a month


```python
df_profit.groupby('month_number')['total_profit'].mean().plot(kind='bar', color = 'orange');
plt.title("Average profit of evry month")
plt.ylabel('Profit')
plt.rcParams['figure.figsize']=(10,8)
plt.xlabel('Sports');
```

# Read all product sales data and show it using a Multiline Plot

```python
fig=plt.figure(figsize=(10,8))
ax1 =plt.subplot(111)
ax1.plot(df.month_number,df.facecream,'o-')
ax1.plot(df.month_number,df.facewash,'v-')
ax1.plot(df.month_number,df.toothpaste,'s-')
ax1.plot(df.month_number,df.bathingsoap,'8-')
ax1.plot(df.month_number,df.shampoo,'>-')
ax1.plot(df.month_number,df.shampoo,'<-')
plt.show()
```

# Read face cream and face wash product sales data and show it using the bar chart

```python
df.groupby('month_number')['facecream'].sum().plot(kind='bar', color = 'green');
plt.title("face-cream sales")
plt.ylabel('Sales')
plt.rcParams['figure.figsize']=(10,8)
plt.xlabel('Month');
```

```python
df.groupby('month_number')['facewash'].sum().plot(kind='bar', color = 'yellow');
plt.title("face-cream sales")
plt.ylabel('Sales')
plt.rcParams['figure.figsize']=(10,8)
plt.xlabel('Month');
```

# Read Bathing soap and face wash data of all months and display it using a Subplot

# Calculate total sale data for last year for each product and show it using a Pie chart
